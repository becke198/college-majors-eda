#Import and Setup

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/college-majors/recent-grads.csv')
df.head()
```

#Clean and Inspect Data

```python
df.info()
df.describe()
df.isnull().sum()
df = df.dropna()
```

#Key Visuals and Insights

```python
sns.scatterplot(x="Unemployment_rate", y="Median", data=df)
plt.title("Median Salary vs. Unemployment Rate")
```

There's a loose negative correlation: Majors with higher median salaries tend to have lower unemployment rates, especially in engineering and technical fields.

```python
top_majors = df.sort_values("Median", ascending=False).head(10)
sns.barplot(x="Median", y="Major", data=top_majors)
plt.title("Top 10 Majors by Median Salary")
```

Engineering dominates the top 10 list, with Petroleum, Chemical, and Computer Engineering leading. These majors offer median salaries above $65,000 at entry-level.

```python
plt.hist(df["Median"], bins=20)
plt.title("Distribution of Median Salaries")
```

Most majors have a median salary between \$35,000 and \$55,000. Only a small number exceed $60,000.


```python
sns.boxplot(x="Major_category", y="Median", data=df)
plt.xticks(rotation=90)
plt.title("Median Salary by Major Category")
```

Engineering, Health, and Business fields not only have higher median salaries but also more consistent outcomes. Arts and Humanities show lower medians and wider variability.

#Summary
